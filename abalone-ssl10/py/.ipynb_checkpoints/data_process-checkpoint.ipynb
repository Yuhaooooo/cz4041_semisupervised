```python
import pandas as pd
import numpy as np
import seaborn as sns

datasetPath='../originalData/'
processedDataStorePath = '../processedData/'
```

## --------------------------
## Get dataset
## --------------------------

```python
dataset='abalone-ssl10'

tra_path = datasetPath + dataset + '-10-' + str(1) + 'tra.dat'
tst_path = datasetPath + dataset + '-10-' + str(1) + 'tst.dat'

with open(tra_path, 'r') as f:
    tra_lines = f.read().splitlines()[1:]
    
with open(tst_path, 'r') as f:
    tst_lines = f.read().splitlines()[1:]
```

```python
def processData(data):
    data = [d for d in data if '@' not in d]
    data2 = []
    for d in data:
#         print(1,d)
        d=d.split(', ')
#         print(2,d)
        if d[8]=='unlabeled':
            d[8]=-1
        else:
            d[8]=int(d[8])
        for i in range(1,8):
            d[i]=float(d[i])
            d[i]=float('%.3f' % d[i])
        data2.append(d)
    df = pd.DataFrame(data2,columns=[1,3,4,5,6,7,8,9,10])
    data_dummies = pd.get_dummies(df.iloc[:,0])
    df2 = pd.concat([data_dummies, df], axis=1, sort=False)
    df2.drop(columns=[1], inplace=True)
    df2 = df2.replace([0,1],[0.0,1.0])
    df2.rename(columns={'F':0,'I':1,'M':2}, inplace=True)
    return df2.iloc[:,:]
```

```python
df_train = processData(tra_lines)
df_test = processData(tst_lines)
```

```python
def correlation_matrix(df):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm

    fig = plt.figure(figsize=(12,12)) # size
    ax1 = fig.add_subplot(111) # 1 graph 1*1 block
    cmap = cm.get_cmap('jet', 30)
    cax = ax1.imshow(df.corr(), cmap=plt.cm.RdYlGn)
    ax1.grid(False)
    plt.title('Correlation Matrix')
    labels=[i for i in range(11)]
    ax1.set_xticks(np.arange(len(labels))) # show all labels
    ax1.set_yticks(np.arange(len(labels)))
    ax1.set_xticklabels(labels,fontsize=12)
    ax1.set_yticklabels(labels,fontsize=12)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax, ticks=[-1,-.5,0,.5,1])
    plt.show()
```

```python
correlation_matrix(df_train)
```

```python
corrFeatureList = []
for i in range(len(df_train.corr())-1):
    for j in range(i+1, len(df_train.corr())-1):
        if abs(df_train.corr().iloc[i,j])>0.5:
            corrFeatureList.append([i,j])
print(corrFeatureList)
```

```python
# coor:3,4,5,6,7,8,9
# coor: 0,1,2
from sklearn.model_selection import train_test_split
setB1,setA1 = train_test_split([0,1,2],test_size=0.5, random_state=10)
setA2,setB2 = train_test_split([3,4,5,6,7,8,9],test_size=0.5, random_state=10)
corrA=setA1+setA2
corrB=setB1+setB2
print(corrA)
print(corrB)
```

```python
df_train_labeled = df_train[df_train[10]!=-1]
df_train_unlabeled = df_train[df_train[10]==-1]
```

```python
X_train_labeled = np.array(df_train_labeled.iloc[:,:-1])
y_train_labeled = np.array(df_train_labeled.iloc[:,-1])

df_trainA_labeled = df_train_labeled[corrA+[10]]
X_trainA_labeled = np.array(df_trainA_labeled.iloc[:,:-1])
y_trainA_labeled = np.array(df_trainA_labeled.iloc[:,-1])

df_trainB_labeled = df_train_labeled[corrB+[10]]  
X_trainB_labeled = np.array(df_trainB_labeled.iloc[:,:-1])
y_trainB_labeled = np.array(df_trainB_labeled.iloc[:,-1])

X_train_unlabeled = np.array(df_train_unlabeled.iloc[:,:-1])

X_test = np.array(df_test.iloc[:,:-1])
y_test = np.array(df_test.iloc[:,-1])
```

```python
# save the data
np.save(processedDataStorePath+'X_train_labeled.npy', X_train_labeled)
np.save(processedDataStorePath+'y_train_labeled.npy', y_train_labeled)
np.save(processedDataStorePath+'X_trainA_labeled.npy', X_trainA_labeled)
np.save(processedDataStorePath+'y_trainA_labeled.npy', y_trainA_labeled)
np.save(processedDataStorePath+'X_trainB_labeled.npy', X_trainB_labeled)
np.save(processedDataStorePath+'y_trainB_labeled.npy', y_trainB_labeled)
np.save(processedDataStorePath+'X_train_unlabeled.npy', X_train_unlabeled)
np.save(processedDataStorePath+'X_test.npy', X_test)
np.save(processedDataStorePath+'y_test.npy', y_test)
```
