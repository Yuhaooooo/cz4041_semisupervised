## testing model

```python
import random
random_list = [random.randint(1,100) for i in range(10)]
print('random seed: ', random_list[0])
from sklearn.model_selection import train_test_split
X_trainVal, X_test, y_trainVal, y_test = train_test_split(X_labeled,y_labeled, test_size=0.97, random_state=random_list[0])

print('train and val: ', len(y_trainVal), ' test: ', len(y_test))

drawNpArray(X_trainVal[:,0],X_trainVal[:,1],y_trainVal)
drawNpArray(X_test[:,0],X_test[:,1],y_test)
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import random

def testXGBoost(X_trainVal,y_trainVal):
    print('total train and val data: ', len(y_trainVal))
    X_train, X_val, y_train, y_val = train_test_split(X_trainVal, y_trainVal, test_size=0.5, random_state=random.randint(1,100))

    drawNpArray(X_train[:,0],X_train[:,1],y_train)
    
#     weights=np.zeros(len(Y_train))
#     weights[Y_train==0]=1
#     weights[Y_train==1]=1

    dtrain = xgb.DMatrix(data = X_train, label = y_train)#, weight=weights)
    Dtest = xgb.DMatrix(data = X_test, label = y_test)
    Dval = xgb.DMatrix(data=X_val, label=y_val)

    param = {'max_depth': 3, 'eta': 1, 'silent': 1, 'objective': 'binary:logistic','estimators':500,'eval_metric':'auc'}

    evallist = [(Dval, 'eval')]
    plst = param.items()

    num_round = 500

    bst = xgb.train(plst, dtrain, num_round, evallist, early_stopping_rounds=10, verbose_eval=50)
    y_pred_score = bst.predict(Dtest)
    y_pred = [round(value) for value in y_pred_score]

    drawNpArray(X_test[:,0],X_test[:,1],y_pred)

    print(confusion_matrix(y_test,y_pred),'\n')
    print(accuracy_score(y_test,y_pred),'\n')
    
    return bst
```

```python
bst=testXGBoost(X_trainVal,y_trainVal)
```

```python
def returnIndexOfPassingProb(data,lowerThreshold,upperThreshold):
    return ([idx for idx, val in enumerate(data) if val>upperThreshold or val<lowerThreshold][::-1])
```

```python
y_pseudo_score = bst.predict(xgb.DMatrix(data=X_unlabeled, label=[1 for i in y_unlabeled]))
print(len(y_pseudo_score),len(y_unlabeled))
lower_bound=.1
upper_bound=.9
print(len(returnIndexOfPassingProb(y_pseudo_score,lower_bound,upper_bound)),' pseudo data are added')
print(len(y_pseudo_score),' unlabeled data in total')
print(len(returnIndexOfPassingProb(y_pseudo_score,lower_bound,upper_bound))/len(y_pseudo_score),'% pseudo data are added')
```

```python
# list(map(int,np.round([.6,.2])))
```

```python
X_testnew = X_trainVal[:]
y_testnew = y_trainVal[:]

bst=testXGBoost(X_testnew,y_testnew)

for i in range(10):
    y_pseudo_score = bst.predict(xgb.DMatrix(data=X_unlabeled, label=[1 for i in y_unlabeled]))
    y_pseudo = list(map(int,np.round(y_pseudo_score)))

    X_new=[]
    y_new=[]
    
    pseudo_label_inverse_index = returnIndexOfPassingProb(y_pseudo_score,lower_bound,upper_bound)
    
    if len(pseudo_label_inverse_index)==0:
        break

    print(len(pseudo_label_inverse_index),' pseudo data are added')
    print(len(y_pseudo),' unlabeled data in total')
    print(len(pseudo_label_inverse_index)/len(y_pseudo),'% pseudo data are added')

    for index in pseudo_label_inverse_index:
        X_new.append(X_unlabeled[index])
        y_new.append(y_pseudo[index])
        X_unlabeled=np.delete(X_unlabeled,index,0) # 0 is row
        y_unlabeled=np.delete(y_unlabeled,index)

    X_new=np.array(X_new)
    y_new=np.array(y_new)

    X_testnew = np.concatenate((X_testnew,X_new),axis=0)
    y_testnew = np.concatenate((y_testnew,y_new),axis=0)

    bst = testXGBoost(X_testnew,y_testnew)
```
