```python
import random
random_list = [random.randint(1,100) for i in range(10)]
print(random_list)
from sklearn.model_selection import train_test_split
X_trainVal, X_test, y_trainVal, y_test = train_test_split(X_labeled,y_labeled, test_size=0.4, random_state=random_list[0])
print('train and val: ', len(y_trainVal), ' test: ', len(y_test))

# df_test = pd.DataFrame({'f1':X_test[:,0],'f2':X_test[:,1],'label':y_test})

# sns.lmplot('f1', 'f2', data=df_test, fit_reg=False,  # x-axis, y-axis, data, no line
#        scatter_kws={"s": 100}, # marker size
#        palette='Set1',
#        hue="label")
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import random

def testXGBoost(X_trainVal,y_trainVal):
    print('total train and val data: ', len(y_trainVal))
    X_train, X_val, y_train, y_val = train_test_split(X_trainVal, y_trainVal, test_size=0.3, random_state=random.randint(1,100))

#     weights=np.zeros(len(Y_train))
#     weights[Y_train==0]=1
#     weights[Y_train==1]=1

    dtrain = xgb.DMatrix(data = X_train, label = y_train)#, weight=weights)
    Dtest = xgb.DMatrix(data = X_test, label = y_test)
    Dval = xgb.DMatrix(data=X_val, label=y_val)

    param = {'max_depth': 3, 'eta': 1, 'silent': 1, 'objective': 'binary:logistic','estimators':500,'eval_metric':'auc'}

    evallist = [(Dval, 'eval')]
    plst = param.items()

    num_round = 500

    bst = xgb.train(plst, dtrain, num_round, evallist, early_stopping_rounds=10)
    y_pred_score = bst.predict(Dtest)
    y_pred = [round(value) for value in y_pred_score]

    df_pred = pd.DataFrame({'f1':X_test[:,0],'f2':X_test[:,1],'label':y_pred})

    sns.lmplot('f1', 'f2', data=df_pred, fit_reg=False,  # x-axis, y-axis, data, no line
           scatter_kws={"s": 100}, # marker size
           palette='Set1',
           hue="label")

    print(confusion_matrix(y_test,y_pred),'\n')
    print(accuracy_score(y_test,y_pred),'\n')
    
    return bst
```

```python
bst=testXGBoost(X_trainVal,y_trainVal)
```

```python
def returnIndexOfPassingProb(data,lowerThreshold,upperThreshold):
    return ([idx for idx, val in enumerate(data) if val>upperThreshold or val<lowerThreshold][::-1])
```

```python
y_pseudo_score = bst.predict(xgb.DMatrix(data=X_unlabeled, label=[1 for i in y_unlabeled]))
print(len(y_pseudo_score),len(y_unlabeled))
```

```python
print(len(returnIndexOfPassingProb(y_pseudo_score,.1,.9)),' pseudo data are added')
print(len(y_pseudo_score),' unlabeled data in total')
print(len(returnIndexOfPassingProb(y_pseudo_score,.1,.9))/len(y_pseudo_score),'% pseudo data are added')
```

```python
X_testnew = X_trainVal[:]
y_testnew = y_trainVal[:]

bst=testXGBoost(X_testnew,y_testnew)

for i in range(5):
    y_pseudo_score = bst.predict(xgb.DMatrix(data=X_unlabeled, label=[1 for i in y_unlabeled]))
    y_pseudo = list(map(int,np.round(y_pseudo_score)))

    X_new=[]
    y_new=[]

    print(len(returnIndexOfPassingProb(y_pseudo_score,0.1,.9)),' pseudo data are added')
    print(len(y_pseudo),' unlabeled data in total')
    print(len(returnIndexOfPassingProb(y_pseudo_score,0.1,.9))/len(y_pseudo),'% pseudo data are added')

    for index in returnIndexOfPassingProb(y_pseudo_score,.1,.9):
        X_new.append(X_unlabeled[index])
        y_new.append(y_pseudo[index])
        X_unlabeled=np.delete(X_unlabeled,index,0) # 0 is row
        y_unlabeled=np.delete(y_unlabeled,index)

    X_new=np.array(X_new)
    y_new=np.array(y_new)
    print('number of newly added: ',len(y_new))

    X_testnew = np.concatenate((X_testnew,X_new),axis=0)
    y_testnew = np.concatenate((y_testnew,y_new),axis=0)

    bst = testXGBoost(X_testnew,y_testnew)
```
